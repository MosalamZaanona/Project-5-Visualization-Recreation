```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Creating a dataframe
df = pd.read_csv('owid-covid-data.csv')
df
```

```python
# We want the location to be one of the following countries: US, UK, Canada, Germany, Italy, and India
Countries = ['United States', 'United Kingdom', 'Canada', 'Germany', 'Italy','India']
df = df[df['location'].isin(Countries)]
df
```

```python
# Now lets create a new dataframe that only contains cloumns that we are concerned about
# We are concerned about location, date, and new_cases_smoothed
df_filtered = df[['location', 'date', 'new_cases_smoothed_per_million']]
df_filtered
```

```python
# Convert date column from string to datetime
df_filtered['date'] = pd.to_datetime(df_filtered['date'])
df_filtered
```

```python
# First we want to extract the unique dates
Days = (df_filtered['date'].unique())
# Now we want to extract the cases for each country seperately
US = df_filtered[df_filtered['location'] == 'United States']['new_cases_smoothed_per_million']
UK = df_filtered[df_filtered['location'] == 'United Kingdom']['new_cases_smoothed_per_million']
Canada = df_filtered[df_filtered['location'] == 'Canada']['new_cases_smoothed_per_million']
Germany = df_filtered[df_filtered['location'] == 'Germany']['new_cases_smoothed_per_million']
Italy = df_filtered[df_filtered['location'] == 'Italy']['new_cases_smoothed_per_million']
India = df_filtered[df_filtered['location'] == 'India']['new_cases_smoothed_per_million']
```

```python
# Now let's work on creating our chart
# We want date to be our x and the new cases of each country to be y

plt.plot(Days[:len(US)], US, label = 'United States')
plt.plot(Days[:len(UK)], UK, label = 'United Kingdom' )
plt.plot(Days[:len(Canada)], Canada, label = 'Canada')
plt.plot(Days[:len(Germany)], Germany, label = 'Germany')
plt.plot(Days[:len(Italy)], Italy, label = 'Italy')
plt.plot(Days[:len(India)], India, label = 'India')

plt.xlabel('Date')
plt.title('Daily new confirmed COVID-19 cases per million people')
plt.xticks([])

plt.legend()
plt.show()
```
